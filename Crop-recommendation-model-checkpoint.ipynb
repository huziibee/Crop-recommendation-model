# INSTALL DEPENDANCIES

```python
import modin.pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearnex import patch_sklearn
from sklearnex.svm import SVC
from sklearnex.linear_model import LogisticRegression
from sklearnex.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

# Patching sklearn with Intel's optimizations

```python
patch_sklearn()
```

# PREPARE THE DATA

```python
file_path = 'Crop_recommendation.csv'
df = pd.read_csv(file_path)

X = df.drop('label', axis=1)
y = df['label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# List to store model names and accuracies

```python
model = []
acc = []
```

# SVM Model & Confusion Matrix for SVM

```python
SVM = SVC(gamma='auto')
SVM.fit(X_train, y_train)
predicted_values = SVM.predict(X_test)
svm_accuracy = accuracy_score(y_test, predicted_values)
print("SVM's Accuracy is: ", svm_accuracy)
print(classification_report(y_test, predicted_values))


cm = confusion_matrix(y_test, predicted_values)
sns.heatmap(cm, annot=True, fmt="d")
plt.title('Confusion Matrix for SVM')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```

# Logistic Regression Model & Confusion Matrix for Logistic Regression

```python
LogReg = LogisticRegression(random_state=2)
LogReg.fit(X_train, y_train)
predicted_values = LogReg.predict(X_test)
logreg_accuracy = accuracy_score(y_test, predicted_values)
print("Logistic Regression's Accuracy is: ", logreg_accuracy)
print(classification_report(y_test, predicted_values))

cm = confusion_matrix(y_test, predicted_values)
sns.heatmap(cm, annot=True, fmt="d")
plt.title('Confusion Matrix for Logistic Regression')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```

# KNN Model & Confusion Matrix for KNN

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
predicted_values = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, predicted_values)
print("KNN's Accuracy is: ", knn_accuracy)
print(classification_report(y_test, predicted_values))

cm = confusion_matrix(y_test, predicted_values)
sns.heatmap(cm, annot=True, fmt="d")
plt.title('Confusion Matrix for KNN')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```

# Decision Tree Model & Confusion Matrix for Decision Tree

```python
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)
predicted_values = dtree.predict(X_test)
dtree_accuracy = accuracy_score(y_test, predicted_values)
print("Decision Tree's Accuracy is: ", dtree_accuracy)
print(classification_report(y_test, predicted_values))

cm = confusion_matrix(y_test, predicted_values)
sns.heatmap(cm, annot=True, fmt="d")
plt.title('Confusion Matrix for Decision Tree')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```

# 🌟 Model Performance Summary 🌟

```python
# Summary of model accuracies
accuracies = {
    "SVM": svm_accuracy,
    "Logistic Regression": logreg_accuracy,
    "KNN": knn_accuracy,
    "Decision Tree": dtree_accuracy
}

# Finding the best performing model
best_model = max(accuracies, key=accuracies.get)
best_accuracy = accuracies[best_model]

# Displaying the accuracies in a flashy way
print("🌟 Model Performance Summary 🌟")
for model, accuracy in accuracies.items():
    print(f"{model} Accuracy: {accuracy:.2%}")

print("\n🏆 Best Performing Model 🏆")
print(f"{best_model} with an outstanding accuracy of {best_accuracy:.2%}!")
```

# LETS TEST THE INPUT VIA THE TERMINAL


```python
def load_dataset(file_path):
    df = pd.read_csv(file_path)
    X = df.drop('label', axis=1)
    y = df['label']
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    return X_scaled, y

def predict_crop(input_values, model_name):
    X_scaled, y = load_dataset('Crop_recommendation.csv')

    if model_name == 'SVM':
        model = SVC(gamma='auto')
    elif model_name == 'Logistic Regression':
        model = LogisticRegression(random_state=2)
    elif model_name == 'KNN':
        model = KNeighborsClassifier(n_neighbors=3)
    elif model_name == 'Decision Tree':
        model = DecisionTreeClassifier(random_state=42)
    else:
        raise ValueError("Invalid model name. Choose from 'SVM', 'Logistic Regression', 'KNN', or 'Decision Tree'.")

    model.fit(X_scaled, y)
    predicted_crop = model.predict([input_values])
    return predicted_crop
```

```python
print("Crop Recommendation Model")
model_name = input("Choose a model for crop recommendation ('SVM', 'Logistic Regression', 'KNN', 'Decision Tree'): ")

input_values = input("Enter values for predicting the crop (space-separated): ").split()
input_values = [float(val) for val in input_values]

predicted_crop = predict_crop(input_values, model_name)
print(f'Predicted Crop: {predicted_crop[0]}')
```
